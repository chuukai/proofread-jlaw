```python
import os
import re
def reversexml(fns):
  logf = open('/content/text.log', 'w')
  writetexts = []
  word = ''
  textxml = ''
  charlist = ''.join(chr(i) for i in range(14,128))
  print("start to loading xml")
#  logf.write(fns + os.linesep)
  textrsnts = open('/content/rsntall/' + fns + '.rsnt', 'r', encoding='utf-8')
  textrsnt = textrsnts.read()
  textrsnts.close()
  ftextxmls = open('/content/xmlall/' + fns + '.xml', 'r', encoding='utf-8')
  textxmls = ftextxmls.read()
  ftextxmls.close()
#  logf.write(textxmls)
  sntstarts1 = [m.end() for m in re.finditer(r"<Sentence>", textxmls)]
  sntstarts2 = [m.end() for m in re.finditer(r"<Sentence .*?>", textxmls)]
  sntstarts3 = [m.end() for m in re.finditer(r"<Sentence [^>]*?/>", textxmls)]
  sntstarts = list(sorted(set(sntstarts1).union(sntstarts2) ^ set(sntstarts3)))
  sntends = [m.start() for m in re.finditer(r"</Sentence>", textxmls)]
  logf.write(",".join(str(n) for n in sntstarts) + os.linesep)
  logf.write(",".join(str(n) for n in sntends) + os.linesep)
#  wordsnt = ''
  wordsnt = textxmls[:sntstarts[0]]
  rsnti = 0
  for snti in range(0, len(sntends)):
#  for snti in range(0, 1):
    wordsnttext = ''
#    print(os.linesep + '%d,' % snti + textxmls[sntstarts[snti]:sntends[snti]] + ',%d' % rsnti)
    for i, textxmlchar in enumerate(textxmls[sntstarts[snti]:sntends[snti]]):
#      print(i, textxmlchar, rsnti, textrsnt[rsnti])
      if rsnti == len(textrsnt):
        snti = len(sntends) - 1
        break
      if textxmlchar == '―':
        wordsnttext += textxmlchar
        rsnti += 1
        continue
      if textxmlchar in charlist:
        wordsnttext += textxmlchar
#        print(i, textxmlchar, rsnti, textrsnt[rsnti] + ' in charlist')
        continue
      if textrsnt[rsnti:rsnti + len('<span>')] == '<span>':
        while True:
#          print(i, textxmlchar, rsnti, textrsnt[rsnti] + ' in span')
          wordsnttext += textrsnt[rsnti]
          rsnti += 1
          if textrsnt[rsnti:rsnti + len('</span>')] == '</span>':
#            print(i, textxmlchar, rsnti, textrsnt[rsnti] + ' end span')
            wordsnttext += '</span>'
            rsnti += len('</span>')
            break
        continue
      if rsnti > len(textrsnt) - 1:
#        print(i, textxmlchar, rsnti, ' rsnti > len(textrsnt) - 1')
#        wordsnt += textxmlchar
        snti = len(sntends) - 1
        break
      if textxmlchar == textrsnt[rsnti]:
#        print(i, textxmlchar, rsnti, textrsnt[rsnti], '%d by equal' % sntends[snti])
        wordsnttext += textxmlchar
        rsnti += 1
#      logf.write(wordsnt + os.linesep)
    rsnti += 1
    wordsnt += wordsnttext
    wordsnttext = ''
    if snti > len(sntends) - 2:
      wordsnt += textxmls[sntends[snti]:]  
      break
    else:
      wordsnt += textxmls[sntends[snti]:sntstarts[snti + 1]]
#    logf.write(wordsnt + os.linesep)
    
#  print(wordsnt)
  word += wordsnt
  wordsnt = ''
  #<span>除く{'あっ': 1}</span>。
  #print(wordsnt)
#  wordxml = open('/content/' + fns + '.rxml', 'w')
#  wordxml.write(word)
#  wordxml.close()
  logf.close()
  return word
#dictsentenceall = {}
#%cd /content/drive/My\ Drive/Colab\ Notebooks/houreibunrui20210510/01-50wakatiall/
#lswakatiall = " ".join(['ls','*.wwakatiall'])
#output = subprocess.getoutput(lswakatiall)
#%cd /content/
#for fni in output.split('\n'):
#  if fni[0:2] in dictsentenceall:
#    dictsentenceall[fni[0:2]].append(fni.strip())
#  else:
#    dictsentenceall.update({fni[0:2] : [fni.strip()]})

#f = open('/content/nmlall/323AC0000000025_20210301_501AC0000000071.sentence.rnml')
#fns = '323AC0000000025_20210301_501AC0000000071'
for bu in range(1,51):
  if bu == 13 or bu == 36:
    f=open('/content/drive/My Drive/Colab Notebooks/houreibunrui20210510/sortuniq/bunruiwall-sort-uniq-%02d-m.csv' % bu)
  else:
    f=open('/content/drive/My Drive/Colab Notebooks/houreibunrui20210510/sortuniq/bunruiwall-sort-uniq-%02d.csv' % bu)
#  f=open('/content/drive/My Drive/Colab Notebooks/houreibunrui20210510/sortuniq/bunruiwall-sort-uniq-%02d.csv' % bu)
  lines=f.readlines()
  f.close()
  writetext = []
  writetexts = []
  outputs = []
  for line in lines:
#  sentencealllists = dictsentenceall[line[0:2]]
#    fns = line[3:52]
    fns = line[3:43]
#fns = '331M50010000018_20210203_503M60000200003'
#if fns:
    print(bu, fns)
    writetexts = reversexml(fns)
    textrstc = open('/content/rxmlall/' + fns + '.rxml', 'w', encoding='utf-8')
    textrstc.write(writetexts)
    textrstc.close()
#  break
#    ziprxml = " ".join(['zip', '/content/drive/My\ Drive/Colab\ Notebooks/houreibunrui20210510/rxmlzip/rxml%02d' % bu, '*.rxml'])
#    print(subprocess.getoutput(ziprxml))
#subprocess.getoutput('rm /content/*.rxml')
```
